```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
filepath = '/content/gdrive/My Drive/'

# 데이터 불러오기
data = pd.read_csv(filepath + 'boston_housing.csv')

# 데이터 전처리: 훈련 데이터와 평가 데이터 분리
train_data = data.iloc[:400, :]
eval_data = data.iloc[400:, :]

# 입력 특징과 타겟 값 추출
X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values
X_eval = eval_data.iloc[:, :-1].values
y_eval = eval_data.iloc[:, -1].values

# 선형 회귀 모델 구축
# 경사 하강법을 사용한 선형 회귀 모델 구현 (하이퍼파라미터 설정 필요)
def gradient_descent(X, y, learning_rate, num_iterations):
    # 초기 가중치 설정
    num_samples, num_features = X.shape
    weights = np.zeros(num_features)

    for i in range(num_iterations):
        # 예측 값 계산
        predictions = np.dot(X, weights)

        # 오차 계산
        error = predictions - y

        # 가중치 업데이트
        gradient = -2 * np.dot(X.T, error) / num_samples
        weights -= learning_rate * gradient

    return weights

# 경사 하강법으로 가중치 학습
learning_rate = 0.001  # 학습률을 더 작게 조정
num_iterations = 10000  # 반복 횟수를 늘림
weights = gradient_descent(X_train, y_train, learning_rate, num_iterations)

# 평가 데이터에 대한 예측
predicted_values = np.dot(X_eval, weights)

# 결과 시각화 (예측 값 vs. 실제 값)
plt.scatter(y_eval, predicted_values) # 예측 값 vs. 실제 값
plt.xlabel("Actual Housing Prices")
plt.ylabel("Predicted Housing Prices")
plt.title("Housing Price Prediction Results")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 파일 읽기
df_data = pd.read_csv('/content/gdrive/My Drive/boston_housing.csv')

# 데이터 준비
X = df_data.drop('medv', axis=1).values
y = df_data['medv'].values
m, n = X.shape

# 하이퍼파라미터 설정
alpha = 0.001
epochs = 10000
initial_weights = np.random.randn(n)
weights = initial_weights.copy()
bias = np.random.randn()

# 경사 하강법으로 선형 회귀 모델 학습
losses = []
for epoch in range(epochs):
    y_pred = X.dot(weights) + bias
    error = y_pred - y
    gradient = -2 * X.T.dot(error) / m
    bias_gradient = -2 * error.sum() / m
    weights -= alpha * gradient
    bias -= alpha * bias_gradient
    mse = np.mean(error**2)
    losses.append(mse)

# 손실 그래프 출력
plt.plot(losses)
plt.xlabel('Epochs')
plt.ylabel('MSE Loss')
plt.title('Loss Over Time')
plt.show()

# 학습된 가중치와 편향 출력
print("학습된 가중치:", weights)
print("학습된 편향:", bias)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 경사 하강법을 위한 시그모이드 함수
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 경사 하강법을 위한 비용 함수 정의
def compute_cost(y, y_pred):
    m = len(y)
    return (-1/m) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))

# 경사 하강법 구현
def gradient_descent(X, y, epochs, learning_rate):
    m, n = X.shape
    theta = np.zeros(n)
    cost_history = []

    for _ in range(epochs):
        z = np.dot(X, theta)
        y_pred = sigmoid(z)
        gradient = np.dot(X.T, (y_pred - y)) / m
        theta -= learning_rate * gradient
        cost = compute_cost(y, y_pred)
        cost_history.append(cost)

    return theta, cost_history

# 데이터 불러오기
df_train = pd.read_csv(filepath + "train.csv")

# 필요한 피쳐 선택 및 전처리
X_train = df_train[["Pclass", "Fare", "Sex", "Age", "SibSp", "Parch", "Embarked"]].copy()
X_train["Sex"] = X_train["Sex"].map({"male": 0, "female": 1})
X_train["Embarked"] = X_train["Embarked"].map({"S": 0, "Q": 1, "C": 2})
X_train["Age"].fillna(X_train["Age"].mean(), inplace=True)
X_train["Fare"].fillna(X_train["Fare"].mean(), inplace=True)

# Feature Scaling (Standardization)
X_train["Fare"] = (X_train["Fare"] - X_train["Fare"].mean()) / X_train["Fare"].std()
X_train["Age"] = (X_train["Age"] - X_train["Age"].mean()) / X_train["Age"].std()

# Target 변수 처리
y_train = df_train["Survived"]

# 훈련 데이터에 절편(intercept)을 추가
X_train["intercept"] = 1

# NumPy 배열로 변환
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()

# 경사 하강법을 사용하여 모델 훈련
epochs = 1000
learning_rate = 0.01
theta, cost_history = gradient_descent(X_train, y_train, epochs, learning_rate)

# 예측을 위한 테스트 데이터 로드 및 전처리
df_test = pd.read_csv(filepath + "test.csv")
X_test = df_test[["Pclass", "Fare", "Sex", "Age", "SibSp", "Parch", "Embarked"]].copy()
X_test["Sex"] = X_test["Sex"].map({"male": 0, "female": 1})
X_test["Embarked"] = X_test["Embarked"].map({"S": 0, "Q": 1, "C": 2})
X_test["Age"].fillna(X_train[:, 3].mean(), inplace=True)
X_test["Fare"].fillna(X_train[:, 1].mean(), inplace=True)
X_test["Fare"] = (X_test["Fare"] - X_train[:, 1].mean()) / X_train[:, 1].std()
X_test["Age"] = (X_test["Age"] - X_train[:, 3].mean()) / X_train[:, 3].std()
X_test["intercept"] = 1
X_test = X_test.to_numpy()

# 생존 예측
y_pred = sigmoid(np.dot(X_test, theta))
y_pred = np.round(y_pred)

# 결과 저장
submission = pd.DataFrame({
    "PassengerId": df_test["PassengerId"].astype(int),
    "Survived": y_pred.astype(int)
})

submission.to_csv(filepath + "result.csv", index=False)
```
